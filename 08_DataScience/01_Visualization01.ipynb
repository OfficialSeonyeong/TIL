- subplot(x,y,z) : x * y 갯수에서 z 번째 위치
- subplot(2,1,1), subplot(2,1,2)

```python
X1 = np.linspace(0.0, 5.0)
Y1 = np.cos(2 * np.pi * X1) * np.exp(-X1)
print(X1)
print(Y1)

X2 = np.linspace(0.0, 2.0)
Y2 = np.cos(2 * np.pi * X2) * np.exp(-X2)
print(X2)
print(Y2)
```

```python
axes01 = plt.subplot(2,1,1)
plt.plot(X1, Y1)
plt.title('subplot 01')

axes02 = plt.subplot(2,1,2)
plt.plot(X2, Y2)
plt.title('subplot 02')

plt.tight_layout() # 자동으로 간격을 맞춰주는 역할
plt.show()
```

```python
axes01 = plt.subplot(1,2,1)
plt.plot(X1, Y1)
plt.title('subplot 01')

axes02 = plt.subplot(1,2,2)
plt.plot(X2, Y2)
plt.title('subplot 02')

plt.tight_layout() # 자동으로 간격을 맞춰주는 역할
plt.show()
```

```python
# 2020 1020
# 이상치 정제
# 구동방식별 고속도로 연비 평균과 도시연비 평균을 극단치를 제외하고 확인
# 각 연비별 이상치 확인 boxplot()

outlier_df = data_df.filter(['hwy','cty'])
outlier_df.boxplot()
plt.show()
```

```python
# 3사분위 정보를 얻어본다면
quantile75 = outlier_df.quantile(q=0.75)
quantile75
```

```python
# 1사분위 정보를 얻어본다면
quantile25 = outlier_df.quantile(q=0.25)
quantile25
```

```python
# IQR (3사분위 - 1사분위)
IQR = quantile75 - quantile25
IQR
```

```python
# 최저 한계치(lower fence)
lower_fence = quantile25 - 1.5*IQR
print(lower_fence)
# 최고 한계치(upper fence)
upper_fence = quantile75 + 1.5*IQR
print(upper_fence)
```

```python
# 극단치 경계 값을 얻어 보자 ( 아래 수평선 )
lower_outlier = outlier_df[outlier_df > lower_fence].min()
print(lower_outlier)

# 극단치 경계 값을 얻어 보자 ( 위 수평선 )
upper_outlier = outlier_df[outlier_df < upper_fence].max()
print(upper_outlier)
```

```python
# 이상치를 결측값으로 변경
hwy_outlier_df = data_df.query('hwy >' +str(upper_outlier['hwy']))
display(hwy_outlier_df)

cty_outlier_df = data_df.query('cty >' +str(upper_outlier['cty']))
display(cty_outlier_df)
```

```python
# copy()
outlier_clean_df = data_df.copy()
outlier_clean_df
```

```python
# hwy 연비에 대한 이상치를 결측값으로 변경
for idx in hwy_outlier_df.index:
    # print(idx)
    outlier_clean_df.loc[idx, 'hwy'] = np.nan
outlier_clean_df
```

```python
# cty 연비에 대한 이상치를 결측값으로 변경
for idx in cty_outlier_df.index:
    # print(idx)
    outlier_clean_df.loc[idx, 'cty'] = np.nan
outlier_clean_df
```

```python
# 결측값 제거한 구동방식별 고속도로, 도시 연비 평균
drv_mean_df = outlier_clean_df.filter(['drv','hwy','cty']).dropna().groupby('drv').mean()
drv_mean_df
```
