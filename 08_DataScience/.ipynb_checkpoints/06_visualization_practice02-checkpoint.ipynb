```python
# 여러 개의 파일을 한꺼번에 읽는 함수
from glob import glob
```

```python
files = glob('./data/oil_data/지역*xls')
files
```

```python
tmp_raw = []
for file in files:
    tmp = pd.read_excel(file, header=2)
    tmp_raw.append(tmp)
tmp_raw
```

```python
oil_station = pd.concat(tmp_raw)
oil_station.head()
```

```python
oil_station.info()
```

```python
# 휘발유 대상으로 분석 진행
stations = pd.DataFrame({'oil_store': oil_station['상호'],
                        'address': oil_station['주소'],
                        'price': oil_station['휘발유'],
                        'self_y_n': oil_station['셀프여부'],
                         'company': oil_station['상표']
                        })
stations
```

```python
# 파생변수 생성 - 구별 주유 가격
stations['direct'] = [addr.split()[1] for addr in stations['address']]
stations.head()
```

```python
stations['direct'].unique()
```

```python
# 서울특별시 -> 성동구, 특별시 -> 도봉구 변경
stations[stations['direct']== '서울특별시']['direct'] = '성동구'
stations[stations['direct']== '특별시']['direct'] = '도봉구'
# 시리즈 형태에서만 변경, 데이터프레임 상에서는 비변경
```

```python
stations[stations['direct']== '특별시']['direct']='도봉구'
stations[stations['direct']== '특별시']['direct']
```

```python
stations.loc[stations['direct']== '특별시','direct']='도봉구'
stations.loc[stations['direct']== '서울특별시','direct']='성동구'
```

```python
stations['direct'].unique()
```

```python
stations[stations['price']== '-']
```

```python
# 가격정보 없는 주유소 정보 삭제, 컬럼 타입 숫자형 변경
stations.loc[stations['price']== '-','price']=np.nan
stations[stations['price']== '-']
```

```python
stations.isnull().sum()
```

```python
stations.dropna(inplace=True)
stations.isnull().sum()
```

```python
# 다른 풀이
stations = stations[stations['price']!= '-']
```

```python
stations['price']=[int(price) for price in stations['price']]
stations.info()
```

```python
# index 처리
stations.reset_index(inplace=True)
stations
```

```python
stations.drop(['index','level_0'], axis=1, inplace=True)
# del stations['index']
```

```python
stations.boxplot(column='price', by='self_y_n')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.set_style('whitegrid')
sns.boxplot(x='company', y='price', hue='self_y_n', data=stations)
sns.swarmplot(x='company', y='price', palette='Set2', hue='self_y_n', data=stations)
```

```python
import json
import folium
import googlemaps
```

```python
# 구별 주유 가격 확인
stations.sort_values(by='price', ascending=False).head()
```

```python
# 피봇 이용하여 구별 가격 정보를 변경, 가격 평균값
stations_pivot = pd.pivot_table(stations,index = 'direct',values='price',aggfunc=np.mean)
stations_pivot
```

```python
# 구별 정보에 대해서 googlmaps 활용해서 시각화
geo_str = json.load(open('./data/02. skorea_municipalities_geo_simple.json', encoding='utf=8'))
```

```python

map = folium.Map(location=[37.55,126.98], zoom_start=10.5)

folium.Choropleth(geo_data = geo_str, data =stations_pivot, columns = [stations_pivot.index, 'price'],
            key_on='feature.id').add_to(map)

map
```

```python
# 주유가격 상위 10개, 하위 10개 주유소를 맵으로 시각화
oil_station_top10 = stations.sort_values(by='price', ascending=False)[:10]
display(oil_station_top10)

oil_station_bottom10 = stations.sort_values(by='price')[:10]
display(oil_station_bottom10)
```

```python
gmap_key = "AIzaSyBu-5obPaXczZKRleHqIAlYctTx1oKEtB0" # 삭제 필요
gmaps = googlemaps.Client(key=gmap_key)
```

```python
lat = []
lng = []
for n in oil_station_top10.index:
    tmp_addr = str(oil_station_top10['address'][n])
    # print(tmp_addr)
    tmp_map = gmaps.geocode(tmp_addr)
    # print(tmp_map)
    tmp_loc = tmp_map[0].get('geometry')

    lat.append(tmp_loc['location']['lat'])
    lng.append(tmp_loc['location']['lng'])
    
print(lat)
print(lng)

oil_station_top10['lat'] = lat
oil_station_top10['lng'] = lng

oil_station_top10
```

```python
for name, lat, lng in zip(oil_station_top10.oil_store, oil_station_top10.lat, oil_station_top10.lng):
    marker = folium.Marker([lat, lng], popup=name, icon=folium.Icon(color='red'))
    marker.add_to(map)
    
map
```
