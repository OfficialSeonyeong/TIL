```python
gender_mean = data[['성별2','일한달의 월 평균임금']]
gender_mean.isna().sum()
gender_mean2 = gender_mean.dropna()
gender_mean2
```

```python
gender_mean2.plot.box()
plt.show()
```

```python
quantile75 = gender_mean2['일한달의 월 평균임금'].quantile(q=0.75)
print(quantile75)
quantile25 = gender_mean2['일한달의 월 평균임금'].quantile(q=0.25)
print(quantile25)
```

```python
upper_fence = quantile75 + 1.5*(quantile75-quantile25)
upper = gender_mean2['일한달의 월 평균임금'][gender_mean2['일한달의 월 평균임금'] < upper_fence].max()
upper
```

```python
outlier = gender_mean2[gender_mean2['일한달의 월 평균임금'] > upper]
print(outlier)

for idx in outlier.index:
    gender_mean2.loc[idx, :] = np.nan
```

```python
gender_mean4 = gender_mean3.groupby('성별2').mean()
gender_mean4
```

```python
# 연령대 계산 파생변수
age = dt.datetime.now().year - data_gender_age['태어난 연도']+1

data_gender_age['연령대'] = (age//10)*10
data_gender_age.head()
```

```python
# 지역별 연령층 분포와 연령층별 지역비율

region_age = data.filter(['태어난 연도','7개 권역별 지역구분'])
region_age['나이'] = dt.datetime.now().year - region_age['태어난 연도'] + 1
region_age.drop('태어난 연도', axis=1, inplace=True)
region_age
```

```python
# 지역코드와 매칭되는 데이터프레임을 만들어서 병합
region_code_df = pd.DataFrame({
    '지역코드': list(range(1,8)),
    '지역명': ['서울','수도권(인천/경기)','부산/경남/울산','대구/경북','대전/충남','강원/충북','경주/전남/전북/제주']
})

region_code_df
```

```python
region_age02 = pd.merge(region_age, region_code_df, 
                        left_on = '7개 권역별 지역구분', right_on = '지역코드', how='outer')
region_age02
```

```python
region_age02.drop('7개 권역별 지역구분', axis=1, inplace=True)
region_age02
```

```python
# 데이터 분석
# 연령층은 나이 구간을 설정 나이 < 30 : 청년층, 나이 < 60: 중년층, 나머지 노년층
# 연령층 파생컬럼 추가

# case 1)

def age(x):
    c = ''
    if x < 30:
        c = '청년층'
    elif x < 60:
        c = '중년층'
    else:
        c = '노년층'
    return c

region_age02['연령층'] = region_age02['나이'].apply(lambda x : age(x))

region_age02
```

```python
# case 2)

# np.select(list, list, defualt) -> Series 리턴
condition01 = [(region_age02['나이'] < 30), 
              (region_age02['나이'] < 60),
              (region_age02['나이'] >= 60)]
condition02 = ['청년층','중년층','노년층']
region_age02['연령층'] =np.select(condition01, condition02)
region_age02
```

```python
region_age02.drop('지역코드',axis=1, inplace=True)
region_age02
```

```python
# 지역과 연령층에 대한 빈도분포를 확인

byregionage = region_age02.groupby(['지역명', '연령층'],as_index=False).count()
byregionage.rename(columns={'나이':'인원'}, inplace=True)
byregionage.head()
```

```python
# 분석결과 pivot 구성, 지역을 인덱스, 연령층을 컬럼으로
bypivot = byregionage.pivot('지역명','연령층','인원')
bypivot
```

```python
bypivot.plot.bar(rot=0)
plt.title('지역별 연령층 분포')
plt.grid()
plt.xlabel('지역')
plt.ylabel('명')
plt.show()
```

```python
fig = plt.figure()
area01 = fig.add_subplot(1,3,1)
area02 = fig.add_subplot(1,3,2)
area03 = fig.add_subplot(1,3,3)

bypivot['청년층'].plot.pie(ax=area01, autopct='%0.1f%%')
area01.title.set_text("청년층")
area01.axis('equal')

bypivot['중년층'].plot.pie(ax=area02, autopct='%0.1f%%')
area02.title.set_text("중년층")
area02.axis('equal')

bypivot['노년층'].plot.pie(ax=area03, autopct='%0.1f%%')
area03.title.set_text("노년층")
area03.axis('equal')

plt.show()
```
