```python
# 1. 각 파일을 pd.read_csv() 로 읽어들인다
# - 헤더 및 구분자 확인요망
ratings = pd.read_csv('C:/Users/A/TIL/08_DataScience/data/ratings.dat', sep='::',header=None)
movies = pd.read_csv('C:/Users/A/TIL/08_DataScience/data/movies.dat', sep='::',header=None)
users = pd.read_csv('C:/Users/A/TIL/08_DataScience/data/users.dat', sep='::',header=None)
```

```python
ratings_df = pd.DataFrame(ratings)
ratings_df.columns = ['userId', 'movieId', 'rating', 'timestamp']
ratings_df
```

```python
movies_df = pd.DataFrame(movies)
movies_df.columns = ['movieId', 'title', 'genres']
movies_df
```

```python
users_df = pd.DataFrame(users)
users_df.drop([3,4], axis=1,inplace=True)
users_df.columns = ['userId', 'gender', 'age']
users_df
```

```python
# rating_df, moives_df -> movieId
# movies_df, users_df -> userId

merge1=pd.merge(ratings_df, movies_df, on='movieId')
merge2=pd.merge(merge1, users_df, on='userId')
merge2
```

```python
# user 평점 갯수 순위 확인

merge2.groupby('userId').size().sort_values().head()
```

```python
user_sort = pd.DataFrame(merge2.groupby('userId').size()).sort_values(by=0, ascending=False)
user_sort
```

```python
m_cnt = pd.DataFrame({'m_cnt':merge2['movieId'].value_counts()})
m_cnt.sort_index()
```

```python
rating_mean_df = merge2.pivot_table('rating','title','gender', aggfunc='mean',fill_value=0)
rating_mean_df
```

```python
# 평점이 200개 이상인 영화
rating = merge2.groupby('title').size()
rating200 = rating.index[rating>=200]
rating200
```

```python
# 평점이 200개 이상인 영화들의 성별 평균 평점
rating200mean = rating_mean_df.loc[rating200]
rating200mean
```

```python
# 여성이 가장 좋아하는 영화 순위
merge2.groupby(['movieId','gender']).count().sort_values('gender')
```
