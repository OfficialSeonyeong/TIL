```python
import seaborn as sns
titanic = sns.load_dataset('titanic') # os error 발생시, engine='python' 추가
```

```python
# 타이타닉 승객 중 나이를 명시하지 않은 고객의 평균 나이 값으로 대체
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
titanic['age']=titanic['age'].astype(int)
titanic

# 타이타닉 승객에 대해 나이와 성별에 의한 age_gender_cat 열을 만들고,
# 조건1. 성별을 나타내는 문자열 male 또는 female로 시작한다.
# 조건2. 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열로 변경
# 조건3. 예시) male27
titanic['age_gender_cat'] = titanic['sex']+titanic['age'].astype(int).apply(str)
titanic
```

##### 실습 : 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.
1. 학생 이름을 나타내는 열을 포함시키지 않고 
데이터프레임 df_score1 을 생성한 후, 
df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
reset_index 명령으로 이 인덱스 열을 명령으로 
일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.

2. 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.

```python
df_score1 = pd.DataFrame({'국어':[80,90,70,80,80],
                     '수학':[70,80,90,70,80],
                     '영어':[80,90,98,78,90]})
df_score1
```

```python
df_score1.index=['김정은','이다영','이예림','김민영','한소연']
df_score1.index.name = '이름'
df_score1
```

```python
df_score2 = df_score1.reset_index()
df_score2
```

```python
df_score2 = df_score2.set_index('이름')
df_score2
```

### DataFrame merge

```python
data1 = {
    '학번' : [1,2,3,4],
    '이름' : ['섭섭해','김한준','김선림','최호진'],
    '학년' : [2,4,1,3]
}

data2 = {
    '학번' : [4,3,2,5],
    '학과' : ['국제학과','국어국문학과','수학학과','수학학과'],
    '학점' : [4.5 ,3.7, 3.4, 4.2]
}
```

```python
stu_df = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
pd.merge(stu_df, major_df, on= '학번', how= 'inner')
```

```python
pd.merge(stu_df, major_df, on= '학번', how= 'left')
```

```python
pd.merge(stu_df, major_df, on= '학번', how= 'right')
```

```python
pd.merge(stu_df, major_df, on= '학번', how= 'outer')
```

```python
pd.merge(stu_df, major_df, left_on = '학번', right_on = '과목코드', how='inner')
```

- 컬럼 인덱스가 아닌 행 인덱스를 기준으로 병합한다면
- left_index, right_index

```python
pop_df01 = pd.DataFrame({
    'city': ['seoul','seoul','seoul','busan','busan'],
    'year': [2000, 2010, 2005, 2020, 2014],
    'pop': [1234567,2345678,3456789,4567890,5678901]
    
})
pop_df01
```

```python
pop_df02 = pd.DataFrame(np.arange(12).reshape(6,2),
    index = [['busan','busan','seoul','seoul','seoul','jinju'],
    [2000, 2010, 2005, 2020, 2014, 2012]],
    columns = ['col01','col02']
    
)
pop_df02
```

```python
pd.merge(pop_df01, pop_df02, left_on=['city','year'], right_index=True)
```

* 다중 인덱스가 아닌 단일 인덱스를 이용하는 방법

```python
data1 = { "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}
```

```python
df01 = pd.DataFrame(data1, index=[1,2,3,4])
df02 = pd.DataFrame(data2, index=[1,2,4,5])
display(df01)
display(df02)
```

```python
mergd_df = pd.merge(df01, df02, left_index = True, right_index=True)
mergd_df
```

```python
# join -> defult : left join
df01.join(df02, how='inner')
```
