- pie chart : 카테고리별 값의 상대적 비교를 해야할 경우
    - 차트의 원형을 유지 : plt.axis('equal')

```python
labels=['짬뽕','순대국밥','비빔밥','육회','햄버거']
sizes = [15,15,20,25,25]
colors = ['gold','blue','red','yellow','green']

plt.title('Pie chart')
plt.pie(sizes, labels = labels, colors=colors, shadow=True, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

- histogram : 구간에 대한 데이터 집계를 시각화할 때

```python
data =  np.random.randn(1000)

plt.title('histogram')
plt.hist(data, bins=20)
plt.show()
```

- scatter plot : 2차원 데이터
    - 데이터 집합의 상관관계

```python
np.random.seed(100)

X = np.random.normal(0,1,100)
Y = np.random.normal(0,1,100)
plt.title('scatter plot')
plt.scatter(X, Y)
plt.show()
```

```python
np.random.seed(100)
X = np.random.rand(30)
Y1 = np.random.rand(30)
Y2 = np.random.rand(30)
Y3 = np.pi * (15*np.random.rand(30))**2

plt.title('bubble chart')
plt.scatter(X, Y1, c= Y2, s=Y3)
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.show()
```

```python
# 그룹 분석으로 각 붓꽃종의 특징값 평균을 구하고 시각화한다면,
iris_mean = iris.groupby('species').mean()
iris_mean.columns.name = 'feature'
iris_mean
iris_mean.plot(kind='bar',rot=0)
```

```python
iris_mean.T.plot.bar(rot=0)
plt.grid(True)
```

- pie, hist, box, scatter

```python
titanic.head()
# 선실별 승객 수 비율을 파이 차트로,
print(type(titanic['pclass'].value_counts()))

titanic['pclass'].value_counts().plot(kind='pie', autopct='%d%%')
plt.axis('equal')
plt.show()
```

#### autopct 사용법
- '%0.2f' — 소숫점 두번째자리 까지 보여준다. 4.08
- '%0.2f%%' — 소숫점 두번째자리까지 보여주고 맨 뒤에 ‘%’ 기호를 붙인다. 4.08%.
- (% 기호를 두 번 쓰는 이유는 문자열 포매팅 할 때 처음 % 기호가 이스케이프 문자처럼 작동하기 때문이다.)
- '%d%%' — 가장 가까운 정수로 반올림하고 맨 뒤에 % 기호를 붙인다.
