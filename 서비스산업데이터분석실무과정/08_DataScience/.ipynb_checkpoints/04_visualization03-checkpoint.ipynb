```python
# 엑셀파일 시트가 여러 개일 때 pd.ExcelFile 사용

elem_school_folium = pd.read_excel('./data/elem_school_folium.xlsx')
elem_school_folium.head()
```

```python
# Feature Selection : 학교명, 학교급구분, 소재지도로명, 위도, 경도
school = elem_school_folium.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
school.head()
```

```python
# 소재지 도로명주소가 서울을 포함하고 학교급구분이 초등학교만 추출
school01 = school[school['학교급구분'].str.contains('초등학교')]
school02 = school01[school01['소재지도로명주소'].str.contains('서울')]
school02.info()
```

```python
# 다른 풀이

# 조건에 부합하는 결과를 데이터 프레임으로 만들기 위해 query()
# condition_df = school.query("학교급구분 == '초등학교' and 소재지도로명주소.str.contains('서울').values")
# condition_df

# index 추출
index_seoul = school['소재지도로명주소'].str.contains('서울특별시')
index_seoul
index_ele = school['학교급구분']=='초등학교'
index_ele

result_df = school.loc[index_seoul & index_ele]
result_df.tail()
```

```python
# 지도 시각화
for name, lat, lng in zip(result_df.학교명, result_df.위도, result_df.경도):
    marker04 = g.Marker([lat, lng], popup=name,icon=g.Icon(color='pink'))
    marker04.add_to(seoul_map)
seoul_map
```

```python
# 데이터 파일 로드
population = pd.read_excel('./data/경기도인구데이터.xlsx',index_col='구분')
population.head()
```

```python
import json
```

```python
# 경계정보를 가진 파일 로드
geo_data = json.load(open('./data/경기도행정구역경계.json', encoding='utf-8'))
geo_data
```

```python
population.columns = population.columns.map(str)

# 지도 생성
Gyeonggi_map = g.Map(location=[37.271121, 127.088539], zoom_start=18,
             tiles='Stamen Terrain')

# 구분된 경계를 표시할 때 사용하는 객체
year = '2017'
g.Choropleth(geo_data = geo_data, data =population[year], columns = [population.index, population[year]],
            key_on='feature.properties.name').add_to(Gyeonggi_map)
Gyeonggi_map
```
