### KMeans 실습

```python
from sklearn.cluster import KMeans

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Wholesale customers dataset
uci_path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv'
sample_df = pd.read_csv(uci_path)
sample_df.head()
```

```python
# 데이터 자료형
sample_df.info()
```

```python
sample_df.describe()
```

```python
# 분석에 사용할 피처를 카피
copy_df = sample_df.iloc[:,:]
copy_df.head()
```

```python
# 표준화 진행
from sklearn.preprocessing import StandardScaler
Standard = StandardScaler()
standard_wholesales = Standard.fit_transform(copy_df)
standard_wholesales
```

```python
# 군집모형 학습 및 예측
# 예측된 결과를 데이터 프레임에 추가
kmeans = KMeans(n_clusters=6, init='k-means++', max_iter=300, n_init=10)
fitted_wholeasles = kmeans.fit(standard_wholesales)
copy_df['cluster_id'] = fitted_wholeasles.labels_
copy_df
```

```python
# 시각화
# plt.scatter(x=copy_df['Grocery'], y=copy_df['Frozen'], c=copy_df['cluster_id'])
# plt.show()

copy_df.plot(kind='scatter', x='Grocery', y='Frozen', c='cluster_id', cmap='Set2', colorbar=True)
plt.show()
```

```python
# 시각화
# plt.scatter(x=copy_df['Milk'], y=copy_df['Delicassen'], c=copy_df['cluster_id'])
# plt.show()

copy_df.plot(kind='scatter', x='Milk', y='Delicassen', c='cluster_id', cmap='Set2', colorbar=True)
plt.show()
```

```python
# PCA 차원축소 
# 군집화 시각화 진행하는 실습

# 표준화 진행
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

Standard = StandardScaler()
standard_wholesales = Standard.fit_transform(copy_df)
wholesales_df =  pd.DataFrame(standard_wholesales)

wholesales_pca = PCA(n_components=2)
wholesales_fitted = wholesales_pca.fit_transform(wholesales_df)

plt.scatter(x=wholesales_fitted[:,0],y=wholesales_fitted[:,1], c=copy_df['cluster_id'])

plt.show()
```
