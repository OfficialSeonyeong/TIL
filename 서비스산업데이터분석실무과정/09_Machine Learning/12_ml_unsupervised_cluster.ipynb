### K-Means Clustering


```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
# dataframe 변환
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target']=iris.target
iris_df
```

- KMeans 객체 생성하고 군집화 수행

```python
iris_means = KMeans(n_clusters=3, random_state=0, init='k-means++', max_iter=300)
iris_means.fit(iris_df.iloc[:,:-1])
```

```python
print(iris_means.labels_)
```

```python
print(iris_means.predict(iris_df.iloc[:,:-1]))
```

```python
# 포인트 별 할당된 그룹을 열로 추가
iris_df['clu_id'] = iris_means.labels_
iris_df
```

```python
# 그룹(target, clu_id)을 통한 sepal_length 빈도 확인
iris_group_cnt = iris_df.groupby(['target','clu_id'])['sepal length (cm)'].count()
iris_group_cnt
```

```python
# 구현 시각화 (PCA 활용)

# 2차원 평면에 데이터 포인트별로 군집화된 결과를 시각화
# 차원숙소(주성분분석)

from sklearn.decomposition import PCA

iris_pca = PCA(n_components=2)
iris_pca_fitted = iris_pca.fit_transform(iris.data)
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(iris_pca_fitted)

iris_df['pca_x'] = iris_pca_fitted[:,0]
iris_df['pca_y'] = iris_pca_fitted[:,1]
iris_df.head()
```

```python
plt.scatter(x=iris_df['pca_x'],y=iris_df['pca_y'], c=iris_df['clu_id'])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c=['r','r','r'], marker='^', s=30)
plt.show()
```

```python
# case2 표준화 통한 차원 축소 및 시각화
Standardization = StandardScaler()
standard_iris = Standardization.fit_transform(iris_df.iloc[:,:4])
df_iris = pd.DataFrame(standard_iris)

iris_pca = PCA(n_components=2)
iris_pca_fitted = iris_pca.fit_transform(df_iris)

iris_df['standard_pca_x'] = iris_pca_fitted[:,0]
iris_df['standart_pca_y'] = iris_pca_fitted[:,1]

plt.scatter(x=iris_df['pca_x'],y=iris_df['pca_y'], c=iris_df['clu_id'])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c=['r','r','r'], marker='^', s=30)

plt.show()
```

#### PCA(차원축소) : Princial Component Analysis
- 데이터의 피처를 압축(테이블 매트릭스의 차원을 낮추는 작업)
- 예) 온도, 습도, 강수량, 미세먼지, 풍속, 태풍 여부, 교통량, 유동 인구 -> 습도와 강수량은 양의 상관관계로 이 툴은 밀접한 연관성이 있다고 볼 수 있다. 이렇게 연관성이 있는 피처들을 하나로 합쳐주는 작업이 주성분 분석


```python
pca_iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
pca_iris_df['target'] = iris.target
pca_iris_df.head()
```

```python
# 각 타겟별 데이터 분포 확인
markers = ['^','o','s']
for i, marker in enumerate(markers):
    # print(i, marker)
    x_data_point = pca_iris_df[ pca_iris_df['target']== i ]['sepal length (cm)']
    y_data_point = pca_iris_df[ pca_iris_df['target']== i ]['sepal width (cm)']
    # print(x_data_point, y_data_point)
    plt.scatter(x_data_point, y_data_point, marker=marker, label=iris.target_names[i])

plt.legend()
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
# 주성분 분석 이전에 컬럼에 대한 데이터 표준화
from sklearn.preprocessing import StandardScaler

# 타겟을 제외한 피처만 추출하여 표준화 작업 진행
Standardazation = StandardScaler()
Standard = Standardazation.fit_transform(pca_iris_df.iloc[:, :-1])
pca_iris_scaler_df = pd.DataFrame(Standard)
pca_iris_scaler_df
```

```python
# 주성분 분석
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
iris_pca = pca.fit(Standard).transform(Standard)

iris_pca_df = pd.DataFrame(iris_pca, columns=['pca01','pca02'])
iris_pca_df['target'] = iris.target
iris_pca_df.describe()
```

```python
# 각 타겟별 데이터 분포 확인

markers = ['^','o','s']
for i, marker in enumerate(markers):
    # print(i, marker)
    x_data_point = iris_pca_df[ iris_pca_df['target']== i ]['pca01']
    y_data_point = iris_pca_df[ iris_pca_df['target']== i ]['pca02']
    # print(x_data_point, y_data_point)
    plt.scatter(x_data_point, y_data_point, marker=marker, label=iris.target_names[i])

plt.legend()
plt.show()
```
