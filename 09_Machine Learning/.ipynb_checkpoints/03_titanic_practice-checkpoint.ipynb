```python
titanic_df = pd.read_csv('./data/titanic_train.csv')
titanic_df.head(3)
```

```python
# 결측값 처리
titanic_df['Cabin'] = np.where(titanic_df['Cabin'].isna(), 'N', titanic_df['Cabin'])
titanic_df['Embarked'] = np.where(titanic_df['Embarked'].isna(), 'N', titanic_df['Embarked'])
age_imputer = SimpleImputer(strategy='mean')
titanic_df['Age'] = age_imputer.fit_transform(titanic_df[['Age']])

# 라벨 인코딩
label_encoder = LabelEncoder()
after_Sex = label_encoder.fit_transform(titanic_df['Sex'])
after_Cabin = label_encoder.fit_transform(titanic_df['Cabin'])
after_Embarked = label_encoder.fit_transform(titanic_df['Embarked'])

titanic_df['Sex']=after_Sex
titanic_df['Cabin']=after_Cabin
titanic_df['Embarked']=after_Embarked
```

```python
# 데이터 세트 만들기
feature = pd.DataFrame([titanic_df['Pclass'],titanic_df['Sex'],titanic_df['Age'],titanic_df['SibSp'],titanic_df['Parch'],titanic_df['Fare'],titanic_df['Cabin'],titanic_df['Embarked']])
feature = feature.T

label = titanic_df['Survived']
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature, label, test_size=0.2)
```

```python
# 의사결정트리 이용한 예측

from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(random_state=20)

dtc.fit(X_train, y_train)
prediction = dtc.predict(X_test)

from sklearn.metrics import accuracy_score
print('예측 정확도 :',accuracy_score(y_test, prediction))
```

```python
# KFold 이용한 예측

from sklearn.model_selection import KFold

kfold = KFold(n_splits=5)
accuracy = []

dtc01 = DecisionTreeClassifier(random_state=20)

for train_idx, test_idx in kfold.split(feature): 
    X_train, X_test = feature.loc[train_idx], feature.loc[test_idx] 
    y_train, y_test = label.loc[train_idx], label.loc[test_idx]
    
    dtc01.fit(X_train, y_train)
    prediction01 = dtc01.predict(X_test)
    accuracy.append(accuracy_score(y_test, prediction01))
print(accuracy)
print(np.mean(accuracy))
```

```python
from sklearn.model_selection import cross_val_score, cross_validate

# cross_val_score 교차검증 예측
dtc02 = DecisionTreeClassifier(random_state=20)
scoring = cross_val_score(dtc02, feature, label, scoring='accuracy', cv=5)
print(scoring)
print(np.mean(scoring))
```

```python
# GridSearchCV (하이퍼 파라미터) 교차검증
X_train, X_test, y_train, y_test = train_test_split(feature, label, test_size=0.2)
dtc03 = DecisionTreeClassifier(random_state=20)

params = {'criterion':['gini', 'entropy'], 'splitter': ['random','best'], 'max_depth': [1,2,3,4,5,6], 'min_samples_split': [2,3]}

grid_gscv_tree=GridSearchCV(dtc03, param_grid=params, cv=20, refit=True)
```

```python
grid_gscv_tree.fit(X_train, y_train)

print('최적의 파라미터: ', grid_gscv_tree.best_params_)
print('높은 정확도 : ', grid_gscv_tree.best_score_)
```

```python
estimator = grid_gscv_tree.best_estimator_
print(estimator)
prediction = estimator.predict(X_test)
print('테스트 세트의 정확도: ',accuracy_score(y_test, prediction))
```
