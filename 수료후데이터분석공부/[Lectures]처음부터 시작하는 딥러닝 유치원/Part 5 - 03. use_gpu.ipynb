# How to use GPU
같은 디바이스 안의 텐서끼리만 연산 가능

```python
import torch
import torch.nn as nn
```

## Convert to CUDA tensor: cuda()
복제해서 해당 디바이스에 저장하는 것으로 기존 변수는 그대로 존재

```python
x = torch.cuda.FloatTensor(2,2)
x
```

```python
x = torch.FloatTensor(2, 2)
x
```

```python
x.cuda()
```

```python
d = torch.device('cuda:0')
```

```python
x.cuda(device=d)
```

```python
x.device
```

## Convert to CUDA tensor: to()
복제해서 해당 디바이스에 저장하는 것으로 기존 변수는 그대로 존재



```python
x.to(device=d)
```

```python
x.device
```

## Covert to CPU tensor from CUDA tensor

```python
x = torch.cuda.FloatTensor(2,2)
```

```python
# 첫번째 방법
x = x.cpu()
```

```python
# 두번째 방법
d =  torch.device('cpu')
x = x.to(d)
```

```python
x.device
```

## Move model from CPU to GPU
Copy가 아니라 Move

```python
def print_params(model):
  for p in model.parameters():
    print(p)
```

```python
linear = nn.Linear(2,2)

print_params(linear)
```

```python
# to GPU
linear = linear.cuda()

print_params(linear)
```

```python
# to CPU
linear = linear.cpu()

print_params(linear)
```

```python
# to GPU
d = torch.device('cuda:0')
linear = linear.to(d)

print_params(linear)
```

```python
# nn.Module은 device 가 없다.
linear.device
```
