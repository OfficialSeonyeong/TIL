## 1. RNN 언어 모델(Recurrent Neural Network Language Model, RNNLM)
* 시점(time step)이라는 개념이 도입된 RNN은 언어 모델을 만들 때 입력의 길이를 고정하지 않아도 된다.
* RNNLM은 기본적으로 예측 과정에서 이전 시점의 출력을 현재 시점의 입력으로 한다.
* 훈련 과정에서는 한 시퀀스의 일부를 입력으로 넣으면 그 뒤에 단어를 예측하도록 훈련된다. 이러한 훈련 기법을 **훈련 강요(teacher forcing)**라고 한다. 즉, 모델이 t 시점에서 예측한 값을 t+1 시점에 입력으로 사용하지 않고, t 시점의 레이블인 실제 알고있는 정답을 t+1 시점의 입력으로 사용한다.
* 훈련 과정에서 이전 시점의 출력을 다음 시점의 입력으로 사용하면서 훈련시킬 수도 있지만 이는 한번 잘못 예측하면 뒤에서의 예측까지 영향을 미쳐 훈련 시간이 느려진다.
* 훈련 과정 동안 출력층에서 사용하는 활성화 함수는 소프트맥스 함수이다. 손실 함수로는 크로스 엔트로피 함수를 사용
