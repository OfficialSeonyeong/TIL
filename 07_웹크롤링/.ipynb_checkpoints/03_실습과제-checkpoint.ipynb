```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.error import HTTPError
from urllib.error import URLError
```

```python
try:
    html = urlopen('https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx')
    
except HTTPError as he :
        print('http error')
except URLError as ue:
    print('url error')
else:
    soup= BeautifulSoup(html.read(),'html.parser')
    print(soup)
```

```python
table = soup.find('table',{'class': 'tData01 tt'})
for tr in table.find_all('tr'):
    print(tr)
```

```python
name_list =[]
AVG_list = []
H_list=[]
HR_list = []
RBI_list = []

for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            name = td.find('a').text
            name_list.append(name)
            AVG = tds[3].text
            AVG_list.append(AVG)
            H = tds[8].text
            H_list.append(H)
            HR = tds[11].text
            HR_list.append(HR)
            RBI = tds[13].text
            RBI_list.append(RBI)
            
```

```python
print(len(name_list), len(AVG_list), len(H_list), len(HR_list), len(RBI_list))
```

```python
import pandas as pd
baseball = pd.DataFrame({
    '이름' : name_list,
    'AVG' : AVG_list,
    'H' : H_list,
    'HR' : HR_list,
    'RBI' : RBI_list
})

baseball
```

```python
baseball.to_csv('baseball_df.csv', mode = 'w', encoding='utf-8')
print('success')
```
