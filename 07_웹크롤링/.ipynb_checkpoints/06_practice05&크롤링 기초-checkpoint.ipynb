# 크롤링이 필요한 이유 & 크롤링 준비 & 이미지 스크래핑 & 자연어 처리

* 이미지 스크랩핑

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.error import HTTPError
from urllib.error import URLError
```

```python
from urllib.parse import quote_plus
```

```python
base_url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query='
keyword = input('검색어를 입력하세요 : ')
imgCnt = int(input('스크랩하려는 이미지 개수 : '))

try:
    url = base_url+quote_plus(keyword)  # quote_plus는 한글을 입력하더라도 url에 알맞게 변환해준다.
    html = urlopen(url)
    
except HTTPError as he :
        print('http error')
except URLError as ue:
    print('url error')
else:
    soup= BeautifulSoup(html.read(),'html.parser', from_encoding='utf-8')
```

```python
img = soup.find_all('img','_img') # url과 제목 등 이미지의 특성이 포함된 태그를 사용해야 함.
img[0]
```

```python
cnt = 1
for i in img:
    imgUrl = i['data-source']
    # print(imgUrl)
    with urlopen(imgUrl) as file:
        with open('./images/'+str(cnt)+'.jpg', 'wb') as imgFile:
            image = file.read()
            imgFile.write(image)
    cnt += 1
    if cnt > imgCnt:
        break

print('Image Download Finished')
```

* 검색키워드를 이용하여 원하는 제목과 링크만 가져오는 실습
> csv 파일로 판다스의 DataFrame을 활용해서 저장해보자

```python
base_url = 'https://search.naver.com/search.naver?where=post&sm=tab_jum&query='
keyword = input('검색어를 입력하세요 : ')

try:
    url = base_url+quote_plus(keyword)
    html = urlopen(url)
    
except HTTPError as he :
        print('http error')
except URLError as ue:
    print('url error')
else:
    soup= BeautifulSoup(html.read(),'html.parser', from_encoding='utf-8')
```

```python
titles = soup.find_all('a','sh_blog_title _sp_each_url _sp_each_title')
titles[0]
```

```python
titles[0]['href']
```

```python
title_list=[]
url_list=[]

for i in titles:
    title_list.append(i.attrs['title'])
    url_list.append(i.attrs['href'])
print(title_list)
print(url_list)
```

```python
print(len(title_list))
print(len(url_list))
```

```python
import pandas as pd
data = pd.DataFrame({'title': title_list, 'url':url_list})
data
```

```python
data.to_csv('python_blog_list.csv', mode='w', encoding='utf-8', index=False)
print('success')
```

# 크롤링을 통한 페이지 처리 방법 익히기
* 크롤러를 코드로 구현

```python
from time import sleep
from random import randint
import pandas as pd
from IPython.core.display import clear_output
from time import time
from bs4 import BeautifulSoup
from requests import get
```

```python
# loop 사이에 잘 진행되고 있는지를 보기위한 indicator sentences
start_time = time() # 현재 시간을 밀리언세컨드로 출력
# start_time 

requests = 0
for _ in range(5):
    requests += 1
    sleep(randint(1,3))
    current_time = time()
    elapsed_time = current_time - start_time
    print("request : {}, frequency : {}".format(requests, requests/elapsed_time))
    clear_output(wait=True) # 결과가 누적되서 보여주는 것이 아니라, 하나 변경되며 보여줌
```

```python
pages = [str(i) for i in range(1,11)]
pages
```

```python
price_list =[]
title_list = []
link_list = []

# loop 사이에 잘 진행되고 있는지를 보기위한 indicator sentences
start_time = time()
requests = 0

for page in pages:
    requests += 1
    url = get('https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery=%ED%8C%8C%EC%9D%B4%EC%8D%AC&pagingIndex='+page+'&pagingSize=40&productSet=total&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&sort=rel&timestamp=&viewType=list')
    sleep(randint(1,3))
    current_time = time()
    elapsed_time = current_time - start_time
    soup= BeautifulSoup(url.text, 'html.parser')
    
    titles = soup.select('.basicList_link__1MaTN')
    title = [title.text for title in titles]
    title_list.extend(title)
    
    prices = soup.select('.price_num__2WUXn')
    price = [price.text for price in prices]
    price_list.extend(price) 
    
    link = [title['href'] for title in titles]
    link_list.extend(link)
```

```python
print(len(title_list))
print(len(price_list))
print(len(link_list))
```

```python
DF = pd.DataFrame({'title':title_list, 'price':price_list, 'link':link_list})
DF
```

```python
DF.to_csv('naver_shopping.csv',mode='w', encoding='utf-8', index=False)
print('success')
```
